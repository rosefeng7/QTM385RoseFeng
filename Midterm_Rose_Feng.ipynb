````python
import random as rd
# Define functions - Please run all of them

def open_file(f):
    file = open(f, 'r')
    lines = file.readlines()
        
    if len(lines)<=5:
        for line in lines:
            print(line, end = "")
    else:
        first_lines = lines[0:5]
        last_lines = lines[-5:] 
        
        for line1 in first_lines:
            print(line1, end = "")
            
        print ('\n...\n')
        
        for line2 in last_lines:
            print(line2, end = "") 
            
    file.close()

def show_map(key):
    letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n',
                   'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    rdletters = letters[:]
    rd.seed(key)
    rd.shuffle(rdletters)
    en_map = dict(zip(letters, rdletters)) 
    de_map = dict(zip(rdletters, letters))
    
    print('\nThe encryption map is:')
    for key, item in en_map.items():
        print(f'{key} => {item}', end = '\t')
    print('\nAnd the decryption map is:')
    for key, item in de_map.items():
        print(f'{key} => {item}', end = '\t')
    print('\n')
    
def encryption(file1, file2, method):
    file = open(file1, 'r')
    words = []
    for line in file:
        for letter in line:
            words.append(letter)
    file.close()
    
    new_l = []
    for i in words:
        if i in method:
            new_l.append(method[i])
        else:
            new_l.append(i)
    file_w = open(file2, 'w')
    file_w.write(''.join(new_l))
    file_w.close()
    
def decryption(file1, method):
    file = open(file1, 'r')
    words = []
    for line in file:
        for letter in line:
            words.append(letter)
    file.close()
    
    de_l = []
    for i in words:
        if i in method:
            de_l.append(method[i])
        else:
            de_l.append(i)
    print('Successfully decrypted.\n')
    print('Your message is: \n')
    print(''.join(de_l))
````

````python
def command():
    fname = 'message.txt'
    encrypted_file = 'encrypted.txt'
    letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n',
                   'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    while True:
        try:
            entry = int(input('Type the encryption key: '))
        except ValueError:
            print('Please type a number between 100 and 999.')
            continue 
        if int(entry) < 100 or int(entry) > 999:
            print('Error: it should be a number between 100 and 999.')
            continue
        else :
            break

    while True:
        x = input('\nEnter your command here: > ')


        if x == 'show-message':
            try: 
                open_file(fname)

            except FileNotFoundError:
                print("Error. " + fname + " is not available. \n")

        elif x == 'set-key':
            while True:
                try:
                    entry = int(input('Type the encryption key: '))
                except ValueError:
                    print('Please type a number between 100 and 999.')
                    continue 
                if int(entry) < 100 or int(entry) > 999:
                    print('Error: it should be a number between 100 and 999.')
                    continue
                else :
                    break

        elif x == 'show-key':
            print('The encryption key is: ' + str(entry) + '\n')

        elif x == 'show-map':
            key = entry
            show_map(key)

        elif x == 'encrypt':
            key = entry
            rdletters = letters[:]
            rd.seed(key)
            rd.shuffle(rdletters)
            en_map = dict(zip(letters, rdletters)) 
            try:
                encryption(fname, encrypted_file, en_map)
                print('Successful encryption.\n')
            except:
                print("Error. The encryption failed.\n")


        elif x == 'decrypt':
            key = entry
            rdletters = letters[:]
            rd.seed(key)
            rd.shuffle(rdletters)
            de_map = dict(zip(rdletters, letters))
            try:
                decryption(encrypted_file, de_map)

            except:
                print("Error. The decryption failed.")

        elif x == 'show-encrypted':
            try:
                open_file(encrypted_file)

            except FileNotFoundError:
                print("Error. 'encrypted.txt' is not available.")  


        elif x == 'bye':
            print('Bye! Have a nice day.')
            break

        else:
            print('Error. Invalid command.')
            continue       
````
